# Data Quality Validation

This notebook demonstrates how booking, fleet, and pricing datasets can be profiled before warehouse publication.

```python
import pandas as pd
from src.common.quality import assert_non_empty, assert_required_columns

bookings = pd.DataFrame([
    {"booking_id": "B1001", "pickup_location_id": "DTW01", "return_location_id": "DTW01", "booked_revenue_amount": 189.50},
    {"booking_id": "B1002", "pickup_location_id": "ORD01", "return_location_id": "ORD01", "booked_revenue_amount": 74.00}
])
bookings
```

```python
checks = [
    assert_non_empty(bookings, "bookings"),
    assert_required_columns(bookings, ["booking_id", "pickup_location_id", "return_location_id", "booked_revenue_amount"]),
]
pd.DataFrame([vars(result) for result in checks])
```

```python
null_summary = bookings.isna().sum().rename("null_count").reset_index().rename(columns={"index": "column_name"})
null_summary
```
